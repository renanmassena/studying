# Classificação Linear

```python
np.random.seed(46)
X1, Y1 = make_classification(n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1)
plt.scatter(X1[:,0], X1[:,1], c=Y1)
```

```python
def plot_model(w1, w2, b):
  plt.scatter(X1[:,0], X1[:,1], c=Y1)
  xmin, xmax = plt.gca().get_xlim()
  ymin, ymax = plt.gca().get_ylim()
  x = np.linspace(xmin, xmax, 50)
  y = (-w1*x - b)/w2
  plt.plot(x, y)
  plt.ylim(ymin, ymax)
  plt.grid(True)

```

```python
w1 = 6
w2 = 5
b = 1.2
```

```python
plot_model(w1, w2, b)
```

```python
def classify(ponto, w1, w2, b):
  res = w1*ponto[0] + w2*ponto[1] + b
  if res >= 0:
    return 1
  else:
    return 0
```

```python
ponto = (1, -2)
yhat = classify(ponto, w1, w2, b)
```

```python
yhat
```

```python
acc = 0
for i, ponto in enumerate(X1):
  yhat = classify(ponto, w1, w2, b)
  if yhat == Y1[i]:
    acc += 1 
```

```python
print('Acurácia: ', acc/len(X1)*100)
```

```python
def plotPerceptron(w1, w2, b):
  plt.scatter(df['petal length (cm)'], df['sepal length (cm)'], c=df['species'])
  xmin, xmax = plt.gca().get_xlim()
  ymin, ymax = plt.gca().get_ylim()
  x = np.linspace(xmin, xmax, 50)
  y = (-w1*x - b)/w2
  plt.plot(x, y)
  plt.ylim(-8, ymax)
  plt.xlim(-8, xmax)
  plt.grid(True)
```

### Multilayer Perceptron

```python
X = df.drop('disease', axis=1)
Y = df['disease']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, stratify=Y, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# x_train = scaler.fit_transform(x_train)
```

```python
scaler.fit(x_train)
```

```python
x_train_std = scaler.transform(x_train)
x_test_std = scaler.transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf = MLPClassifier(random_state=42, hidden_layer_sizes=(6,), activation='logistic')
clf.fit(x_train_std, y_train)
```

```python
yhat = clf.predict(x_test_std)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
clf.score(x_train_std, y_train)
```

```python
accuracy_score(yhat, y_test)
```

```python
confusion_matrix(y_test, yhat)
```

```python
conf = confusion_matrix(y_test, yhat)
```

```python
sns.heatmap(pd.DataFrame(conf), annot=True)
plt.xlabel('P R E D I C T')
plt.ylabel('R E A L')
plt.show()
```

### Tuning

```python
# degrau
# sigmoide
# tanh
# relu
```

```python
clf = MLPClassifier(random_state=42, 
                    hidden_layer_sizes=(6,), 
                    activation='tanh',
                    alpha=0.0001,
                    max_iter=800,
                    learning_rate_init=0.001,
                    )

clf.fit(x_train_std, y_train)
yhat = clf.predict(x_test_std)

print('Acc Train: ', clf.score(x_train_std, y_train))
print('Acc Test: ', accuracy_score(yhat, y_test))

conf = confusion_matrix(y_test, yhat)
sns.heatmap(pd.DataFrame(conf), annot=True)
plt.xlabel('P R E D I C T')
plt.ylabel('R E A L')
plt.show()
```

```python
clf = MLPClassifier(random_state=42, 
                    hidden_layer_sizes=(6,6,), 
                    activation='tanh',
                    alpha=0.0001,
                    max_iter=800,
                    learning_rate_init=0.001,
                    )

clf.fit(x_train_std, y_train)
yhat = clf.predict(x_test_std)

print('Acc Train: ', clf.score(x_train_std, y_train))
print('Acc Test: ', accuracy_score(yhat, y_test))

conf = confusion_matrix(y_test, yhat)
sns.heatmap(pd.DataFrame(conf), annot=True)
plt.xlabel('P R E D I C T')
plt.ylabel('R E A L')
plt.show()
```
